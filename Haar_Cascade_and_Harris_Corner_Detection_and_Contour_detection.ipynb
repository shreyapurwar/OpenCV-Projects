Face Detection uSing Open Cv


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread("multi_faces.jpg")
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,10))
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(20,10))
plt.imshow(gray,cmap="gray")
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
faces = face_cascade.detectMultiScale(gray, 4,6)
```

```python
img_with_detections=np.copy(image)
```

```python
for(x,y,w,h)in faces:
  cv2.rectangle(img_with_detections,(x,y),(x+w,y+h),(255,0,0),5)
plt.figure(figsize=(20,10))
plt.imshow(img_with_detections)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('waffle.jpg')
image_copy =np.copy(image)
image_copy=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_copy)
```

```python
#Detect Corners
```

```python
gray=cv2.cvtColor(image_copy,cv2.COLOR_BGR2GRAY)
```

```python
gray=np.float32(gray)
dst = cv2.cornerHarris(gray, 2,3,0.04)#2 means it's making 2 by 2 pixel square and 3 is size of sobel operator and 0.04 determines which is a corner here
dst=cv2.dilate(dst,None)
plt.imshow(dst,cmap='gray')
```

```python
#Performing operation called Dilation kind of filtering
```

```python
#performing Hough Transformation Now let's extarct and displat the strong corners
```

```python
thresh=0.001*dst.max()
corner_image=np.copy(image_copy)
for j in range(0,dst.shape[0]):
  for i in range(0,dst.shape[1]):
    if(dst[j,i]>thresh):
      cv2.circle(corner_image,(i,j),1,(0,255,0),1)
plt.imshow(corner_image)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
image = cv2.imread('thumbs_up_down.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
retval,binary = cv2.threshold(gray,225,255,cv2.THRESH_BINARY_INV)
plt.imshow(binary,cmap='gray')
```

```python
#find the contours
```

```python
contours,hierarchy = cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
#Draw the contours
```

```python
contours_image=np.copy(image)
contours_image = cv2.drawContours(contours_image,contours,-1,(0,255,0),3)
plt.imshow(contours_image)
```
